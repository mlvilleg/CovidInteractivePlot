```python
import requests
from bs4 import BeautifulSoup



```

```python
URL = "https://realpython.github.io/fake-jobs/"
page = requests.get(URL)


soup = BeautifulSoup(page.content, "html.parser") #parse the data

results = soup.find(id="ResultsContainer") # find that specific HTML element by its ID:
print(results.prettify())


```

```python


job_link = "https://"    
job_link += input("What website are you using?") #Get the job
job_link += "/jobs/" #add the query parameter

job_type = input("What type of position are you looking for?")
if ' ' in job_type:
    #spit the word
    job_type = job_type.split()
    job_link += job_type[0] + "%20" + job_type[1]
else:
    job_link += job_type






get_page = requests.get(job_link) #See if the page returns successfully
parse_link =  BeautifulSoup(get_page.content, "html.parser")  #parse the page











print(parse_link.prettify())


```
